### Analiza podatkov

```python
#kateri avtor ima najboljšo povprečno skupno oceno na skupni listi? (skupna lista je lista vseh treh osnovnih list)
skupna_avtorji = pd.concat([lista1, lista2, lista3]).join(avtorji, on = 'id_avt')
skupna_avtorji.groupby('avtor').mean().sort_values('ocena_skupaj', ascending=False)[:10]['ocena_skupaj']
```

```python
#kako priljubljenost pada med prvimi desetimi avtorji?
skupna_avtorji.groupby('avtor').mean().sort_values('ocena_skupaj', ascending=False)[:10]['ocena_skupaj'
                                                                                        ].plot().set_ylabel("skupna ocena")
```

```python
#kateri avtor ima največ knjig na skupni listi?
skupna_avtorji.groupby('avtor').size().sort_values(ascending=False)
```

```python
#kateri avtor ima največ knjig na skupni listi? - graf za prvih 50 knjig
skupna_avtorji.groupby('avtor').size().sort_values(ascending=False)[:50].plot(kind='bar').set_ylabel("število knjig")
```

```python
#kateri naslov se je uvrstil na vse tri liste?
skupna_naslovi = pd.concat([lista1, lista2, lista3]).join(naslovi, on = 'id_nasl')
skupna_naslovi.groupby('naslov').size().sort_values(ascending=False)
```

```python
#Stone Soup ima več ponovitev kot obstaja list, ker je več avtorjev napisalo knjigo s tem naslovom:
lista3.join(naslovi, on = 'id_nasl').join(avtorji, on = 'id_avt')[
    lista3.join(naslovi, on = 'id_nasl').join(avtorji, on = 'id_avt')['naslov'] == 'Stone Soup']
```

```python
#ko razvrstimo knjige glede na avtorje in naslove, dobimo željen seznam:
skupna_naslovi.join(avtorji, on = 'id_avt').groupby(['naslov', 'avtor']).size().sort_values(ascending=False)
```

```python
#tabela naslovov, ki so se uvrstili na vse tri liste:
st_ponovitev = skupna_naslovi.join(avtorji, on = 'id_avt').groupby(['avtor','naslov']).count()
st_ponovitev[st_ponovitev.id_nasl == 3]
```

```python
#prvih pet avtorjev po številu knjig znotraj posamezne liste:
display(lista1.join(avtorji, on = 'id_avt')['avtor'].value_counts().head())
display(lista2.join(avtorji, on = 'id_avt')['avtor'].value_counts().head())
display(lista3.join(avtorji, on = 'id_avt')['avtor'].value_counts().head())
```

```python
#najboljša skupna ocena glede na avtorje znotraj posamezne liste:
l = lista1.join(avtorji, on = 'id_avt').groupby('avtor').mean().sort_values('ocena_skupaj', ascending=False).head()
display(l[['ocena_skupaj']])
l2 = lista2.join(avtorji, on = 'id_avt').groupby('avtor').mean().sort_values('ocena_skupaj', ascending=False).head()
display(l2[['ocena_skupaj']])
l3 = lista3.join(avtorji, on = 'id_avt').groupby('avtor').mean().sort_values('ocena_skupaj', ascending=False).head()
display(l3[['ocena_skupaj']])
```

```python
#10 najslabše ocenjenih knjig skupne liste
slabe=skupna_naslovi.groupby('naslov').mean().sort_values('ocena_skupaj', ascending=False).tail(10).join(avtorji, on = 'id_avt')
slabe[['ocena', 'st_glasov', 'ocena_skupaj', 'avtor']]
```

```python
#generiramo novo listo s knjigami, ki se pojavijo na vseh treh listah in izračunamo njihovo povprečno skupno oceno
skupna = skupna_naslovi.join(avtorji, on = 'id_avt').set_index(['avtor', 'naslov'])
st_ponovitev = skupna_naslovi.join(avtorji, on = 'id_avt').groupby(['avtor','naslov']).count()
ustrezna = st_ponovitev[st_ponovitev.mesto == 3].reset_index(level=['avtor','naslov'])[['avtor', 'naslov']]
ustrezna.join(skupna, on=['avtor', 'naslov']).groupby('naslov').mean()[
    ['ocena','st_glasov', 'ocena_skupaj']].sort_values('ocena_skupaj', ascending=False).reset_index(level=['naslov'])


```

```python
#generiramo novo listo glede na število glasov
ustrezna.join(skupna, on=['avtor', 'naslov']).groupby('naslov').mean()[
    ['ocena','st_glasov', 'ocena_skupaj']].sort_values('st_glasov', ascending=False)
```

```python
#kako pa so knjige razvrščene glede na oceno?
ustrezna.join(skupna, on=['avtor', 'naslov']).groupby('naslov').mean()[
    ['ocena','st_glasov', 'ocena_skupaj']].sort_values('ocena', ascending=False)
```

```python
# Na spletni strani, kjer smo zajeli podatke, je navedeno, da je skupna ocena odvisna od več parametrov, 
# tudi od števila glasov in ocene. Tu sta grafa, ki kažeta odvisnost št. glasov od uvrščenega mesta
# (glede na novo listo, ki vsebuje le knjige, ki so se pojavile na vseh treh listah) in odvisnost 
# ocene od uvrščenega mesta.
a = ustrezna.join(skupna, on=['avtor', 'naslov']).groupby('naslov').mean()[
    ['ocena','st_glasov', 'ocena_skupaj']].sort_values('ocena_skupaj', ascending=False).reset_index(level=['naslov'])
fig, axs = plt.subplots(1,2)
a1 = a['ocena'].plot(ax=axs[0])
a2 = a['st_glasov'].plot(ax=axs[1])
a1.set_xlabel("mesto")
a2.set_xlabel("mesto")
a1.set_ylabel("ocena")
a2.set_ylabel("št. glasov")
```

```python
#ta graf prikazuje odvisnodt skupne ocene od uvrščenega mesta.
g = a['ocena_skupaj'].plot()
g.set_xlabel("mesto")
g.set_ylabel("ocena skupaj")
```

```python
# Recimo, da skupno oceno dobimo tako, da pomnožimo št. glasov in oceno in ju množimo z neko konstanto. 
# Kako dobro se tak graf prilega resničnemu izračunu skupne ocene?
k = ustrezna.join(skupna, on=['avtor', 'naslov']).groupby('naslov').mean()[
    ['ocena','st_glasov', 'ocena_skupaj']].sort_values('ocena_skupaj', ascending=False).reset_index(level=['naslov'])

k['vrednost'] = k.st_glasov * k.ocena * (1/200)
#k['st_glasov'] = k['st_glasov'].apply(lambda x: x*(1/100))
pd.concat((k[['vrednost']], a[['ocena_skupaj']]), axis = 1).plot().set_xlabel("mesto")
```

```python
#zanima nas, pri katerem mestu na listi št. glasov močno vpade? (primerjamo vse tri liste)
o = lista1[['mesto', 'st_glasov']].rename(index=str, columns={"st_glasov": "št. glasov 1"})
o1 = pd.merge(o, lista2[['mesto', 'st_glasov']].rename(index=str, columns={"st_glasov": "št. glasov 2"}))
o2 = pd.merge(o1, lista3[['mesto', 'st_glasov']].rename(index=str, columns={"st_glasov": "št. glasov 3"})).set_index(['mesto'])
o2.plot()
```

```python
#katere so knjige, ki imajo vsaj 500000 glasov in oceno manjšo od 4.0?
u = ustrezna.join(skupna, on=['avtor', 'naslov']).groupby('naslov').mean()
u[(u['st_glasov'] >= 500000) & (u['ocena'] < 4.0)][['ocena','st_glasov','ocena_skupaj']]
```
